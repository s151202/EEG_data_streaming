```python
import pyglet
import time
import datetime
from random import randint
import random
import numpy as np
```

```python
def make_sprite(img, window, center = 1):
    sprite = pyglet.sprite.Sprite(img)
    if center:
        sprite.x = (window.width - img.width)//2
        sprite.y = (window.height - img.height)//2
    return sprite   
```
